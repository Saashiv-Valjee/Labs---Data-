## Pseudorandom numbers
A simple pseudorandom number generator follows the formula: 

$$r_{i+1} = (ar_i + c) \mod M,$$ 

where $a$, $c$, and $M$ are specified integer constants. This gives integers between 0 and $M-1$. 

This is called a *linear congruential random number generator*.  It generates a string of apparently random positive integers, by iterating the equation over and over.

Clearly you need an initial value to begin iterating through the sequence.  This value $r_0$ is called the *'seed'* of the pseudorandom number generator. 

**Write a function to generate random integers using the relation above.**

*Hint:* In Python, the modulo (“remainder”) operator is `%`.

Start by using the values $M = 2^6$, $c = 1$, $a = 57$, and begin iterating with a seed of $r_0 = 10$.

**Determine the period of this generator:** that is, how many values does it generate before returning to the starting point $r = 10$, or initial seed value?

```python
r_values=[10]
M=2**6
c=1
a=57
count=0


for i in range(100):
    
    
    
    r_i= (a*r_values[i]+c) % M
    count=count+1
    
    r_values.append(r_i)
    if r_i == 10:
        print('Period:',count-1)
        break
    else:
        continue

print(r_values)
```

## Interlude: more on `array`s

We’re going to need two useful `array` functions in the following code. (Formally speaking, one is a function and one a *method*: note the slight differences in syntax between how we call them.) 

The `sum` method by default adds up every number in an array, but can be restricted to do so along a particular axis:

```python
y = np.array([[1, 2, 3],#6
              [4, 5, 6]])#15
              #5,7,9
print("Sum all:")
print(y.sum())#adds all the elements

print("Sum along columns:")#
print(y.sum(axis=0))

print("Sum along rows:")
print(y.sum(axis=1))
```

The `cumsum` (“cumulative sum”) function adds up the first number, the first two numbers, the first three numbers, and so on. Again, by default this will be done over every element, but we can choose to do so only along a particular axis:

```python
print("Cumulative sum of all elements:")
print(np.cumsum(y))

print("Cumulative sum along columns:")
print(np.cumsum(y, axis=0))

print("Cumulative sum along rows:")
print(np.cumsum(y, axis=1))
```

## Random walks
Random numbers can be used to sample or explore a parameter space in an unbiased way.  Here we will examine the class of behaviour known as “random walks”.

We consider a “walker” that takes a step of constant length $L=1$, but random directions. We want to determine the path taken after a large number of steps $N$.  This process is a good analogue for Brownian motion or other diffusive collisonal processes (*e.g.*, the movement of a photon from the core of a star to the photosphere, diffusion of perfume molecules through a room, and so forth).

**Complete the function below** that generates the $x$ and $y$ coordinates for points on a random walk with step length 1. We calculate these as follows:

1. Choose a random direction for each step: *i.e.*, an angle in the range $[0, 2\pi)$.
1. Calculate the corresponding steps in the $x$ and $y$ directions, using trigonometry:

1. Use the `cumsum()` function to add up these steps, so that the $n$th entry represents the coordinates after $n$ steps have been taken.

```python
def random_walk_2d(nsteps, iseed):
    """ This function computes a 2D random walk from the origin
            with n steps of uniform length 1 
            
        Inputs:    
            nsteps - number of steps in random walk
            iseed - seed for random number generator
            
        Returns (total_x, total_y)
            total_x - x coordinates of walk at each step (np.array)
            total_y - y coordinates of walk at each step (np.array) """

    seed(iseed)          # initialise random number generator
    a=0
    b=2*np.pi

    theta = random(nsteps)*2*np.pi         # choose a random number between 0 and 2*pi
    
    step_x = np.cos(theta) # calculate the x and y steps from trigonometry
    step_y = np.sin(theta)         # using theta and the diagram above
    
    total_x = np.cumsum(step_x)   # use cumsum to transform from a list of steps 
    total_y = np.cumsum(step_y)   # to a list of points visited in the random walk
    
    return total_x, total_y
```

Simple theoretical arguments indicate that averaging over many walks the root-mean-of-the-square of the distance from the starting point (origin in this case) should scale as the following (see,
$$
\langle R(N)^2 \rangle^{1/2} \simeq \sqrt{N} L_{rms}
$$
where $L_\text{rms}$ is the root-mean-of-the-square of the step size.  In our case, because each step is the same size, $L_\text{rms}=L=1$.

**Using a cell below, adapt your code from above to compute:**
1. The square of the distance from the origin at all steps $N$ in an individual walk (e.g., $R(N)^2 = x(N)^2 + y(N)^2$)
2. The mean of the squares of the distance from the origin at step $N$ taking the average over all $K$ walks,
    $$
        \langle R(N)^2 \rangle = \frac{1}{K}\sum_{k=1}^K R_{(k)}(N)^2
    $$
    where $K$ is the number of walks and $k$ runs over them, and
3. The square root of the mean of the square of the distance over all runs at step $N$
$$
R_\text{rms}(N) = \sqrt{ \frac{\langle R(N)^2 \rangle}{K} }
$$
4. Plot $R_\text{rms}(N)$ vs. $N$ on a log-log plot and confirm the $N^{1/2}$ relationship (perhaps by plotting the theoretical result on the same axes).

Note: In step (2) above it may be useful to recall that array arithmetic is done element by element. If you encounter difficulty using this in your code, ask a demonstrator for some suggestions to get started.


```python
from numpy.random import random, seed
nsteps = 10000
nruns = 100
plt.figure()
# calculate the rms distance after N steps
nruns_list=[5,10,50,100]
for nruns in nruns_list:
    squares = []
    for K in range(nruns):
        x,y =random_walk_2d(nsteps,seed())
        squares.append(x**2+y**2)
    rms = np.sqrt(sum(squares)/nruns**2)
    
    xplot = np.arange(1, nsteps +1)
    plt.loglog(xplot,rms,label='rms over {:2d} runs'.format(nruns)) 
nruns_list = [5,10,50,100]
    
    
plt.loglog(xplot,np.sqrt(xplot),label='theoretical curve')
plt.ylabel(r"${\langle R^2 \rangle}^{1/2}$")
plt.xlabel('$N$')
plt.legend()
```
